```python
import os  
import shutil  
  
# parent directory  
parent_directory = '/home/susmoh-3/r7020_cvip/data/real_and_decoy/raw/test'  

# List of subfolder names  
subfolders = [  
    #'camera_color_camera_info',  
    'camera_color_image_raw',  
    #'camera_depth_camera_info',  
    #'camera_depth_image_raw',  
    'camera_depth_points',  
    #'camera_left_ir_image_raw',  
    #'camera_right_ir_image_raw',  
    #'depth_registered_colored_pointclouds'  
]  
  
 
target_folder = os.path.join(parent_directory, '2_color_images_depth_points')  
os.makedirs(target_folder, exist_ok=True)  

count = 0  
# Iterate over each subfolder  
for folder in subfolders:  
    # Get the full path of the subfolder  
    folder_path = os.path.join(parent_directory, folder) 
     
    # Check if the subfolder exists  
    if os.path.exists(folder_path): 
        # Iterate over all files in the subfolder  
        for filename in os.listdir(folder_path):  
            file_path = os.path.join(folder_path, filename)  
              
            # Check if it's a file (not a directory)  
            if os.path.isfile(file_path):  
                # Copy the file to the target folder  
                shutil.copy(file_path, target_folder) 
                count += 1
  
print("files copied to the 'all_files' folder:", count)  
```

```python
# List to store filenames and timestamps  
files_with_timestamps = []  
  
# Iterate over all files in the directory  
for filename in os.listdir(target_folder):  
    file_path = os.path.join(target_folder, filename)  
      
    # Check if it's a file  
    if os.path.isfile(file_path):  
        # Extract the timestamp from the filename  
        try:  
            # Assuming the format is some_text_timestamp.extension  
            parts = filename.rsplit('_', 1)  
            timestamp_part = parts[1].split('.')[0]  
            timestamp = int(timestamp_part)  
            files_with_timestamps.append((timestamp, filename))  
        except (IndexError, ValueError):  
            continue  
  
# Sort the files by timestamp  
files_with_timestamps.sort()  
  
# Rename the files with index as prefix  
for index, (timestamp, filename) in enumerate(files_with_timestamps):  
    new_name = f"{index}_{filename}"  
    new_file_path = os.path.join(target_folder, new_name)  
      
    # Rename the file  
    os.rename(os.path.join(target_folder, filename), new_file_path)  
    print(f"Renamed: {filename} to {new_name}")  
```
